```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import numpy as np
```

```python
#Read the data
df = pd.read_csv("50_Startups.csv", encoding= "latin1")
df.head()
```

```python
# rename the data for your comfort
df1 = df.rename({'R&D Spend':'RD','Administration':'Ad','Marketing Spend':'MS'},axis=1)
df1
```

```python
# Droping State column 
df1.drop(["State"],inplace=True,axis = 1)
```

```python
df1.info()
```

```python
#check for missing values
df1.isna().sum()
```

# Preparing a model

```python
#Build model
import statsmodels.formula.api as smf 
model = smf.ols('Profit~RD+Ad+MS',data=df1).fit()
```

```python
#Coefficients
model.params
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

# Calculating VIF

```python
rsq_RD = smf.ols('RD~Ad+MS',data=df1).fit().rsquared  
vif_RD = 1/(1-rsq_RD)

rsq_Ad = smf.ols('Ad~RD+MS',data=df1).fit().rsquared  
vif_Ad = 1/(1-rsq_Ad) 

rsq_MS = smf.ols('MS~RD+Ad',data=df1).fit().rsquared  
vif_MS = 1/(1-rsq_MS) 

# Storing vif values in a data frame
d1 = {'Variables':['RD','Ad','MS'],'VIF':[vif_RD,vif_Ad,vif_MS]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

# Residual Analysis

Test for Normality of Residuals (Q-Q Plot)

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

# Residual Plot for Homoscedasticity

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

# Detecting Influencers/Outliers

# Cooks Distance

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(df1)), np.round(c,3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#index and value of influencer where c is more than .5
(np.argmax(c),np.max(c))
```

# Improving the model

```python
df1
```

```python
#Discard the data points which are influencers and reasign the row number (reset_index())
df2=df1.drop(df1.index[[46]],axis=0).reset_index()
```

```python
#Drop the original index
df2=df2.drop(['index'],axis=1)
```

```python
df2
```

# Build Model

```python
while np.max(c)>0.5 :
    model=smf.ols('Profit~Ad+MS+RD',data=df2).fit()
    (c,_)=model.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    df2=df2.drop(df2.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    df2
else:
    final_model=smf.ols('Profit~Ad+MS+RD',data=df2).fit()
    final_model.rsquared , final_model.aic
    print("Thus model accuracy is improved to",final_model.rsquared)
```
